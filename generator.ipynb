```python
directory_name = "output"
f1 = open("coords.txt","r")
coordinates = f1.read().split("\n")
flag=True
with open('names-courses-completion.csv', mode ='r')as file:
  csvFile = csv.reader(file)
  for lines in csvFile:
      list_info = lines
      #print(lines)
      name_to_print= list_info[0]
      course_name_to_print= list_info[1]
      date_to_print= list_info[2]+ " " + list_info[3]
      image = cv.imread("myimage_500.png")  
      cv2_im_rgb = cv.cvtColor(image,cv.COLOR_BGR2RGB)  
      pil_im = Image.fromarray(cv2_im_rgb)  
      draw = ImageDraw.Draw(pil_im)  
      font = ImageFont.truetype("./fonts/G_ari_bd.TTF", 34)
      font1 = ImageFont.truetype("./fonts/G_ari_bd.TTF", 40)
      font2 = ImageFont.truetype("./fonts/arial.TTF", 26)
      draw.text((int(coordinates[0]), int(coordinates[1])), name_to_print, font=font , fill='#2b2828')
      draw.text((int(coordinates[2]), int(coordinates[3])), course_name_to_print , font=font1, fill='#2b2828')
      draw.text((int(coordinates[4]), int(coordinates[5])), date_to_print , font=font2, fill='#2b2828') 
      cv2_im_processed = cv.cvtColor(np.array(pil_im), cv.COLOR_RGB2BGR) 
      type(cv2_im_processed)
      if flag:
        cv.imshow('Certificate', cv2_im_processed) #Shows sample image
        cv.imwrite('./output/'+name_to_print+'.png',cv2_im_processed)
        #os.startfile('output.png')
        cv.waitKey(0)  
    
        cv.destroyAllWindows()
      
```
